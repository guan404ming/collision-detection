```python
import csv
import matplotlib.pyplot as plt

def print_data(file):
    # Read the CSV file
    data = []
    with open(file, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            data.append(int(row[0]))

    # Separate odd and even rows into separate lists
    odd_values = data[::2]
    even_values = data[1::2]

    # Create a line plot to compare the values
    x = range(1, len(odd_values) + 1)  # x-axis values
    plt.plot(x, odd_values, label='Basic')
    plt.plot(x, even_values, label='QuadTree')
    plt.xlabel('Matrix Size')
    plt.ylabel('Compare Time')
    plt.title('Comparison of QuadTree and Basic')
    plt.legend()
    plt.show()
```

```python
print_data('./output/output.csv')
```
